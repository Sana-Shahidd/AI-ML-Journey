```python
# Ask the user to input a number and convert it to integer. Handle ValueError.
# so first understand what is value error. Function error is when function argument is of inappropriate type. int(a)
a=input("Enter an integer: ")
try:
    b=int(a)
    print(b)
except ValueError:
    print("you are trying to do wrong type casting.")
except Exception as e:
    print(e.with_traceback)
```

```python
# Open a file named data.txt and print its content. Handle FileNotFoundError.
try:
    with open("data.txt","r") as f:
        f.read()
except FileNotFoundError:
    print("Sorry!File not found.")
except Exception as e:
    print(e.with_traceback)
```

```python
# Write a program that takes a list of numbers and asks the user for an index to access. Handle IndexError.
a=[1,2,3,4,5]
try:
    access_index=int(input("Enter the index that you want to access the value of: "))
    print(a[access_index])
except IndexError:
    print("index out of range")
except Exception as e:
    print("unexpected error: ",e)
```

```python
# Handle KeyError while accessing a key in a dictionary entered by the user.
# keyerror is you are trying to access a key that is not present.
dic={
    "apple":2,
    "mango":4,
    "banana":5
}
try:
    key = input("Enter the key you want to access: ")
    print("Value:", dic[key])
except KeyError:
    print("Such key is not preent in dictionary.")
except Exception as e:
    print("unexpected error: ",e)
```

```python
# Create a custom exception called AgeTooLowError. Ask the user for their age and raise this excepti
# on if age < 18.
# Create a custom exception
class AgeTooLowError(Exception):
    pass

try:
    age = int(input("Enter your age: "))

    if age < 18:
        raise AgeTooLowError("Age must be 18 or above.")

    print("Access granted. You are eligible.")

except ValueError:
    print("Invalid input. Please enter a valid number.")

except AgeTooLowError as e:
    print("Custom Exception Caught:", e)

finally:
    print("Program finished.")
```
