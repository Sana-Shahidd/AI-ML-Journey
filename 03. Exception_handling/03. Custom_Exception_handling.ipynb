Create a custom exception class NegativeNumberError.
Write a program that takes a number from the user and raises NegativeNumberError if the number is negative. Otherwise, print the square of the number.

Example:

Input: -5
Output: NegativeNumberError: Negative numbers are not allowed

```python
class NegativeNumberError(Exception):
    def __init__(self, message="Negative numbers are not allowed"):
        super().__init__(message)

try:
    num = int(input("Enter a number: "))

    if num < 0:
        raise NegativeNumberError()

    print("Square =", num ** 2)

except NegativeNumberError as e:
    print("NegativeNumberError:", e)

except ValueError:
    print("Please enter a valid integer.")
```

Library mangement system

```python
import os


# -------------------- Custom Exceptions --------------------

class DuplicateBookError(Exception):
    def __init__(self, message="Book with this ID already exists."):
        super().__init__(message)


class BookNotFoundError(Exception):
    def __init__(self, message="Book not found."):
        super().__init__(message)


class OutOfStockError(Exception):
    def __init__(self, message="Book is out of stock."):
        super().__init__(message)


# -------------------- Library Class --------------------

class Library:
    def __init__(self, filename="books.txt"):
        self.filename = filename

        # Create file automatically if it does not exist
        if not os.path.exists(self.filename):
            with open(self.filename, "w") as file:
                pass

    def load_books(self):
        books = []

        with open(self.filename, "r") as file:
            for line in file:
                line = line.strip()
                if line:  # skip empty lines
                    parts = line.split(",")
                    book = {
                        "id": parts[0],
                        "name": parts[1],
                        "author": parts[2],
                        "copies": int(parts[3])
                    }
                    books.append(book)

        return books

    def save_books(self, books):
        with open(self.filename, "w") as file:
            for book in books:
                file.write(f"{book['id']},{book['name']},{book['author']},{book['copies']}\n")

    def add_book(self, book_id, name, author, copies):
        books = self.load_books()

        for book in books:
            if book["id"] == book_id:
                raise DuplicateBookError()

        new_book = {
            "id": book_id,
            "name": name,
            "author": author,
            "copies": copies
        }

        books.append(new_book)
        self.save_books(books)
        print("Book added successfully.")

    def borrow_book(self, book_id):
        books = self.load_books()

        for book in books:
            if book["id"] == book_id:
                if book["copies"] == 0:
                    raise OutOfStockError()

                book["copies"] -= 1
                self.save_books(books)
                print(f"Book borrowed successfully: {book['name']}")
                return

        raise BookNotFoundError()

    def return_book(self, book_id):
        books = self.load_books()

        for book in books:
            if book["id"] == book_id:
                book["copies"] += 1
                self.save_books(books)
                print(f"Book returned successfully: {book['name']}")
                return

        raise BookNotFoundError()

    def display_books(self):
        books = self.load_books()

        if not books:
            print("No books available in the library.")
            return

        print("\nLibrary Books:")
        print("-" * 50)
        for book in books:
            print(f"ID: {book['id']}")
            print(f"Name: {book['name']}")
            print(f"Author: {book['author']}")
            print(f"Copies: {book['copies']}")
            print("-" * 50)


# -------------------- Main Program --------------------

library = Library()

while True:
    print("\n===== Library Management System =====")
    print("1. Add Book")
    print("2. Borrow Book")
    print("3. Return Book")
    print("4. Display Books")
    print("5. Exit")

    choice = input("Enter your choice: ")

    try:
        if choice == "1":
            book_id = input("Enter book ID: ")
            name = input("Enter book name: ")
            author = input("Enter author name: ")
            copies = int(input("Enter number of copies: "))

            library.add_book(book_id, name, author, copies)

        elif choice == "2":
            book_id = input("Enter book ID to borrow: ")
            library.borrow_book(book_id)

        elif choice == "3":
            book_id = input("Enter book ID to return: ")
            library.return_book(book_id)

        elif choice == "4":
            library.display_books()

        elif choice == "5":
            print("Exiting program...")
            break

        else:
            print("Invalid choice. Please try again.")

    except DuplicateBookError as e:
        print("DuplicateBookError:", e)

    except BookNotFoundError as e:
        print("BookNotFoundError:", e)

    except OutOfStockError as e:
        print("OutOfStockError:", e)

    except ValueError:
        print("Invalid input. Copies must be a number.")

    except Exception as e:
        print("Unexpected Error:", e)
```
