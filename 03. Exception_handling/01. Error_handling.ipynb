# Why is it important to handle exceptions
# how to handle exceptions
# -> Try except block
*user experience
*security breaches can occur


```python
#try except block
#Jo code garbar kr sakta ha wo try me uska solution except me
#lets create a file
with open('sample.txt','w') as f:
    f.write("Hello World")
#try catch demo
try:
    with open("sample1.txt","r") as f:
        print(f.read())
except:
    print("Sorry file not found")
#being an exxperienced programmmer you must write the code in try except when you are dealing with external resource
```

```python
#catching a specific exception
try:
    f= open('sample.txt',"r")
    print(f.read())
    print(f)
    print(5/0)
except Exception as e:
    print(e.with_traceback)
```

```python
#catching a specific exception
try:
    f= open('sample.txt',"r")
    print(f.read())
    print(f)
    print(5/2)
    L=[1,2,3]
    L[100]
except FileNotFoundError:
    print("File not found error")
except NameError:
    print("Variable  not declared")
except ZeroDivisionError:
    print("cannot divide by zero")
except Exception as e:
    print(e)
#generic error must be at the end otherwise this dominate all others
```

```python
#else block
try:
    f= open('sample.txt',"r")
except FileNotFoundError:
    print("File not found")
except Exception as e:
    print(e)
else:
    print(f.read()) #you are supersure ke yh thk ha jb try bilkul sae se chale ga tb else me jai ga
finally:
    print("yh to print hoga hi..") #no matter rasta konsa ha
    f.close()
    print("File closed successfully")
```
