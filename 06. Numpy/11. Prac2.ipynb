```python
# Q-4: Your are given an array which is havig some nan value. 
# You job is to fill those nan values with most common 
# element in the array.
arr=np.array([[1,2,np.nan],[4,2,6],[np.nan,np.nan,5]])
# np.modf(arr) #np.modf() splits numbers into fractional and integer parts, which does not help with NaN replacement.
values = arr[~np.isnan(arr)]
mode=np.bincount(values.astype(int)).argmax() #bincount works on integers so argmax return the highest value
arr[np.isnan(arr)] = mode
print(arr)
```

```python
# Q-5: Write a NumPy program
# to find the missing data in a given array. Return a boolean matrix.
# also try to fill those missing values with 0. For that, you can use np.nan_to_num(a)
# import numpy as np

# np.array([[3, 2, np.nan, 1],
#           [10, 12, 10, 9],
#           [5, np.nan, 1, np.nan]])
a=np.array([[3, 2, np.nan, 1],
          [10, 12, 10, 9],
          [5, np.nan, 1, np.nan]])
a[np.isnan(a)]=0
print(a)
```

```python
import numpy as np

a = np.array([[3, 2, np.nan, 1],
              [10, 12, 10, 9],
              [5, np.nan, 1, np.nan]])

# 1. Boolean matrix for missing values
missing_data = np.isnan(a)
print("Missing data (boolean matrix):")
print(missing_data)

# 2. Fill missing values with 0
filled_array = np.nan_to_num(a, nan=0)
print("\nArray after filling NaN with 0:")
print(filled_array)
```

```python
# Q-7: Plot this below equation.
# y=ex−e−xex+e−x 

# Note: This equation is called tanh activation function. In deep learning, many times this function is used. If you find some difference between the sigmoid function and this tanh function, note that to your notebook.
import matplotlib.pyplot as plt
x = np.linspace(-5,5,100)
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

# plot
plt.plot(x,y)
plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("tanh(x)")
```

| Feature   | Sigmoid           | Tanh                           |
| --------- | ----------------- | ------------------------------ |
| Formula   | (1/(1+e^{-x}))    | ((e^x-e^{-x})/(e^x+e^{-x}))    |
| Range     | 0 to 1            | -1 to 1                        |
| Center    | Not zero-centered | Zero-centered                  |
| Use in DL | Older networks    | Often better for hidden layers |

```python
# Q-9: Write a program implement Boradcasting Rule to check if two array can be added or not.
# Given tuples of shapes.

# shape of a- (3,2,2)
# shape of b- (2,2)

def can_broadcast(shape_a, shape_b):

    a = list(shape_a[::-1])
    b = list(shape_b[::-1])

    for i in range(max(len(a), len(b))):
        
        dim_a = a[i] if i < len(a) else 1
        dim_b = b[i] if i < len(b) else 1

        if dim_a != dim_b and dim_a != 1 and dim_b != 1:
            return False

    return True


shape_a = (3,2,2)
shape_b = (2,2)

print("Broadcast possible:", can_broadcast(shape_a, shape_b))
```
