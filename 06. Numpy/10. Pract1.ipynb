```python
# Q-11: Softmax function
# Create a Python function to calculate the Softmax of the given numpy 1D array. The function only accepts the numpy 1D array, otherwise raise error.

# σ(z⃗ )i=ezi∑Kj=iezj 

# https://en.wikipedia.org/wiki/Softmax_function

# Example 1
# Input:
# [86.03331084 37.7285648  48.64908087 87.16563062 38.40852563 37.20006318]
# Output:
# [2.43733249e-01, 2.56112115e-22, 1.41628284e-17, 7.56266751e-01, 
#        5.05514197e-22, 1.50974911e-22]
# Example 2:
# Input:
# [33.17344305 45.61961654 82.05405781 80.9647098  68.82830233 91.52064278]
# Output:
# [4.57181035e-26, 1.16249923e-20, 7.73872596e-05, 2.60358426e-05,
#        1.39571531e-10, 9.99896577e-01]
import numpy as np
def softmax(arr):
    if not isinstance(arr, np.ndarray):
        raise TypeError("Input must be a NumPy array")
    if arr.ndim!=1:
        raise ValueError("Input must be a 1D NumPy array")
    exp_values = np.exp(arr)
    softmax_values = exp_values / np.sum(exp_values)
    
    return softmax_values

a = np.array([86.03331084, 37.7285648, 48.64908087, 87.16563062, 38.40852563, 37.20006318])
print(softmax(a))
```

```python
# Q-13: Dates
# Create a python function named date_array that accepts two dates as string format and returns a numpy array of dates between those 2 dates. The function only accept 2 strings, 
# otherwise raise error. The date format should be like this only: 2022-12-6. 
# The end date should be included and for simplicity, choose dates from a same year.

# Example 1:

# Input:

# date_array(start = '2020-09-15', end = '2020-09-25')
# Output:

# ['2020-09-15', '2020-09-16', '2020-09-17', '2020-09-18',
#  '2020-09-19', '2020-09-20', '2020-09-21', '2020-09-22',
#  '2020-09-23', '2020-09-24', '2020-09-25']
# Example 2:

# Input:

# date_array(start = '2022-12-01', end = '2022-12-06')
# Output:

# ['2022-12-01', '2022-12-02', '2022-12-03', '2022-12-04', '2022-12-05', '2022-12-06']
# Example 3:

# Input:

# date_array(start = '2020-11-25', end = '2020-11-30')
# Output:

# ['2020-11-25', '2020-11-26', '2020-11-27', '2020-11-28',
#  '2020-11-29', '2020-11-30']

import numpy as np

def date_array(start, end):

    # type checking
    if not isinstance(start, str) or not isinstance(end, str):
        raise TypeError("Both inputs must be strings")

    start_date = np.datetime64(start)
    end_date = np.datetime64(end)

    # include end date
    dates = np.arange(start_date, end_date + np.timedelta64(1, 'D'))

    return dates

print(date_array('2020-09-15', '2020-09-25'))
```

```python
# Q-18 Answer below asked questions on given array:
# Fetch Every alternate column of the array
# Normalise the given array
# https://en.wikipedia.org/wiki/Normalization_(statistics)

# There are different form of normalisation for this question use below formula.

# Xnormalized=X−XminXmax−Xmin 

# arr1=np.random.randint(low=1, high=10000, size=40).reshape(8,5)
# Given
# arr1=np.random.randint(low=1, high=10000, size=40).reshape(8,5)
# arr1
import numpy as np

# given array
arr1 = np.random.randint(low=1, high=10000, size=40).reshape(8,5)

print("Original Array:\n", arr1)

#Fetch every alternate column
alternate_columns = arr1[:, ::2]
print("\nAlternate Columns:\n", alternate_columns)

#Normalize the array
xmin = np.min(arr1)
xmax = np.max(arr1)

normalized_array = (arr1 - xmin) / (xmax - xmin)

print("\nNormalized Array:\n", normalized_array)
```

```python
# Q-19: Write a function which will accept 2 arguments.
# First: A 1D numpy array arr

# Second: An integer n {Please make sure n<=len(arr)}

# Output: The output should be the nth largest item out of the array

# Example1 : arr=(12,34,40,7,1,0) and n=3, the output should be 12
# Example2 : arr=(12,34,40,7,1,0) and n=1, the output should be

import numpy as np

def find_value(arr, n):
    if n > len(arr):
        print("n should be <=", len(arr))
        return None
    
    arr_sorted = np.sort(arr)[::-1]   # sort descending
    return arr_sorted[n-1]


a = np.array([12,34,40,7,1,0])
result = find_value(a,3)

print(result)
```
