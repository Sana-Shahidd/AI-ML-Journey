Broadcasting Rules

1. Compare shapes from right to left.
2. Dimensions must either:
   - be equal, OR
   - one of them must be 1.
3. If dimension is 1, it expands to match the other dimension.
4. If neither dimension is 1 and they are not equal → broadcasting fails.

Example:
(3,1) and (1,3) → (3,3)

```python
import numpy as np
a = np.array([[1],
              [2],
              [3]])      # shape (3,1) then (3,3)

b = np.array([10,20,30]) # shape (3,) ->(3,1) ->(3,3)
print(a+b) 
```

```python
import numpy as np
a = np.array([1,2,3]) #(3,)
b = 5 

print( a + b)
print() #[1+5 ,2+5, 3+5]

a1 = np.arange(6).reshape(2,3)  # (2,3)
b1 = np.array([10,20,30])       # (,3) -->> (1,3) --->>(2,3)
print(a1+b1)
print()

a2 = np.arange(6).reshape(2,3)
b2 = np.array([[10],
               [20]])  # (2,1) -->> (2,3)
print(a2+b2)
print()

A = np.ones((3,4))
B = np.array([10,20,30,40])

print( A + B)
```

```python
a3 = np.arange(12).reshape(4,3)
b3 = np.array([1,2,3]) # (,3) -->> (1,3) -->>> (4,3)

print(a3 + b3 )
print()
```

```python
a5 = np.arange(12).reshape(3,4)  # (3,4)
b5 = np.arange(3) # (,3) -->> (1,3) -->> ( 3, 3 ) shapes didnt match so won't broadcast 
#                  we can only increament 1
print(a5)
print("\n", b5)

print( a5 + b5 )
```
