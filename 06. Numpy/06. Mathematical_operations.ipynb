Scalar operations

```python
import numpy as np
```

```python
d=np.array([1,2,3,4,5])
print(d+1)
print(d-1)
print(d*1)
print(d//1)
print(d/1)
print(d%2)
```

Relational Operators

```python
#return true and false based on conditions
d=np.array([1,2,3,4,5])
print(d>1)
print(d<1)
print(d!=1)
print(d<=1)
print(d>=1)
print(d==1)
```

Vector operations

```python
a=np.array([1,2,3,4,5])
b=np.array([3,4,5,6,7])
print("Sum",a+b)
print("Mul",a-b)
print("Addition",a*b)
print("Subtraction",a/b)
```
