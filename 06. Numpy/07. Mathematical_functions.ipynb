```python
import numpy as np
#all operations are element wise
a=np.array([1,2,3,4,5])
b=np.array([5,6,7,8,9])
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
print(np.power(a,2))
print(np.abs(a))
print(np.mod(a,b))
print(np.exp(a))
print(np.log10(a))
print(np.sqrt(a))
```
