```python
# i. Sort this matrix. np.sort()
# ii. Sort this matrix based on values in 2nd column.
# iii. Sort this matrix based on max value in each row.
# iv. Sort based on elements value.
import numpy as np

a = np.random.randint(0,100,12).reshape(3,4)

print("Original Matrix:\n",a)

# i
print("\nRow wise sorted:\n",np.sort(a))

# ii
print("\nSorted by 2nd column:\n",a[a[:,1].argsort()])

# iii
print("\nSorted by row max:\n",a[a.max(axis=1).argsort()])

# iv
print("\nAll elements sorted:\n",np.sort(a,axis=None))
```

```python
import numpy as np

a = np.random.randint(0, 75, 20).reshape(5, 4)
print("Original array:\n", a)

# i. Add marks for one extra subject to all existing students
extra_subject = np.array([[23],
                          [24],
                          [25],
                          [26],
                          [12]])
a = np.hstack((a, extra_subject))
print("\nAfter adding extra subject:\n", a)

# ii. Add two new students (now each must have 5 subject marks)
new_students = np.array([[34, 45, 56, 67, 21],
                         [11, 22, 33, 44, 55]])
a = np.vstack((a, new_students))
print("\nAfter adding two new students:\n", a)

# iii. Add total marks column
total = np.sum(a, axis=1).reshape(-1, 1)
a = np.hstack((a, total))
print("\nAfter adding total marks column:\n", a)

# iv. Sort by total marks in descending order and show top 2 rows
sorted_array = a[a[:, -1].argsort()[::-1]]
top2 = sorted_array[:2]
print("\nTop 2 students based on total marks:\n", top2)
```

```python
# Q-7
import numpy as np

np.random.seed(400)
arr = np.random.randint(100, 1000, 200).reshape((1, 200))

print("Original array:\n", arr)

min_val = arr.min()
max_val = arr.max()

print("\nMinimum value:", min_val)
print("Maximum value:", max_val)

# Remove min and max
filtered = arr[(arr != min_val) & (arr != max_val)]
print("\nArray after removing min and max:\n", filtered)
```

```python
# Q-8
np.random.seed(42)
arr = np.random.randint(100, 701, 20)

print("Original array:\n", arr)

# Limit values between 100 and 200
clipped = np.clip(arr, 100, 200)
print("\nClipped array:\n", clipped)

# Sort the array
sorted_arr = np.sort(clipped)
print("\nSorted array:\n", sorted_arr)

# Cumulative sum
cumulative = np.cumsum(sorted_arr)
print("\nCumulative sum:\n", cumulative)
```

```python
# Q-9
np.random.seed(42)
arr = np.random.rand(20)

print("Original array:\n", arr)

# Round to 3 decimal places
rounded = np.round(arr, 3)
print("\nRounded array:\n", rounded)

# 0th percentile and minimum
p0 = np.percentile(rounded, 0)
min_val = rounded.min()

# 100th percentile and maximum
p100 = np.percentile(rounded, 100)
max_val = rounded.max()

print("\n0th percentile:", p0)
print("Minimum value :", min_val)
print("Are they equal?", p0 == min_val)

print("\n100th percentile:", p100)
print("Maximum value   :", max_val)
print("Are they equal?", p100 == max_val)

# Difference between 51st and 50th percentile
p51 = np.percentile(rounded, 51)
p50 = np.percentile(rounded, 50)

print("\n51st percentile:", p51)
print("50th percentile:", p50)
print("Difference:", p51 - p50)
```
