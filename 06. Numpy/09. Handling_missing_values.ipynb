```python
import numpy as np
a = np.array([1,2,3,np.nan,4,5,np.nan])
print(a)
mask=np.isnan(a)
print(mask)
```

```python
cleaning_data=a[~np.isnan(a)] #~np.isnan(a) this gives non nan values
print(cleaning_data)
```

```python
# replacing the missing values with mean
data = np.array([1, 2, np.nan, 4, 5, np.nan])
print(data)
mean_val=np.nanmean(data)
print("mean", mean_val)
data[np.isnan(data)] = mean_val
print("After replacing NaN:", data)
```

```python
filled_data_zero = np.nan_to_num(data, nan = 0) # replace with zero
print("Data after filling with 0:", filled_data_zero)
```

Ploting Graph
we can plot graph of every function using Numy and Matplotlib
Graph Type	Purpose / Use	Basic Syntax
Line Graph	Show trend or relationship between two continuous variables	plt.plot(x, y)
Scatter Plot	Show individual data points, correlation between variables	plt.scatter(x, y)
Bar Graph	Compare quantities across categories (vertical bars)	plt.bar(categories, values)
Horizontal Bar	Compare quantities across categories (horizontal bars)	plt.barh(categories, values)
Histogram	Show distribution/frequency of a dataset	plt.hist(data)
Pie Chart	Show proportion of categories as parts of a whole	plt.pie(sizes, labels=labels)
Multiple Lines	Compare multiple trends in the same graph	plt.plot(x, y1); plt.plot(x, y2)
Area Plot / Fill	Highlight area under a curve	plt.fill_between(x, y1, y2)
Stacked Bar	Compare multiple quantities stacked on top of each other	plt.bar(categories, values1); plt.bar(categories, values2, bottom=values1)
