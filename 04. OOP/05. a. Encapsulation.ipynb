Collections of objects

```python
#list of objects
class Person:
    def __init__(self,name,gender):
        self.name=name
        self.gender=gender

    def __str__(self):
        return f"{self.name} ({self.gender})"
    
    def __repr__(self):
        return f"Person('{self.name}', '{self.gender}')"
    
p1=Person("sana","female")
p2=Person("saba","female")
p3=Person("sara","female")
L=[p1,p2,p3]
print(L)
```

Getter and setter using decorators..

```python
#The setter name must match the property name. 
# If your getter method is with adc name then settre should be with abc and decorator will be @adc.setter
class Student:
    def __init__(self, cgpa):
        self.__cgpa = cgpa

    @property
    def cgpa(self):
        return self.__cgpa

    @cgpa.setter
    def cgpa(self, value):
        if 0 <= value <= 4:
            self.__cgpa = value
        else:
            print("Invalid CGPA")
```
