Question 1: Vehicle Management System (Abstract Class + Abstract Methods)

Create an abstract class Vehicle using ABC.

Requirements:

The class must contain two abstract methods:
start_engine()
fuel_type()
The class must also contain a normal method:
display_info() that prints "This is a vehicle".

Create two child classes:
Car
Bike
Each class must implement both abstract methods.
Create objects of both classes and call all methods.
Try creating an object of Vehicle and explain why it fails.

Example Output (idea):

This is a vehicle
Car engine started
Fuel type: Petrol

This is a vehicle
Bike engine started
Fuel type: Petrol

Concepts Covered:
Abstract class
Abstract methods
Method implementation
Restriction on abstract class objects

```python
from abc import ABC, abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass
    
    @abstractmethod
    def fuel_type(self):
        pass
    
    #normal method
    def display_info(self):
        print("This is a vehicle")


class Car(Vehicle):
    def start_engine(self):
        print("Car engine started.")
    
    def fuel_type(self):
        print("Fuel type is: Petrol.")
    
    #normal method
    def display_info(self):
        super().display_info()

class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine started.")
    
    def fuel_type(self):
        print("Fuel type is: Petrol.")
    
    #normal method
    def display_info(self):
        super().display_info()

vehicle=[Car(), Bike()]
for s in vehicle:
    s.display_info()
    s.start_engine()
    s.fuel_type()
```
