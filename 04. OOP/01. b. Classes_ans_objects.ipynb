```python
# Q-1: Rectangle Class
# Write a Rectangle class in Python language, allowing you to build a rectangle with length and width attributes.
# Create a Perimeter() method to calculate the perimeter of the rectangle and a Area() method to calculate the area of ​​the rectangle.
# Create a method display() that display the length, width, perimeter and area of an object created using an instantiation on rectangle class.
```

```python
class Rectangle: #class  
    # Encapsulation means binding data and methods together inside a class. not complete hiding but concept there

    def __init__(self, length, width): #constructor
        self.length = length  #instance variable. They belong to each object separately.
        self.width = width

    def perimeter(self):  #instance method 1
        return (2*(self.length+self.width))

    def area(self):    #instance method 2 , belongs to class. Operate on object data using self
        return self.length*self.width 
    
    def display(self):
        print("The length of rectangle is: {}".format(self.length))
        print("The width of rectangle is: {}".format(self.width))
        print("The area of rectangle is: {}".format(self.area()))
        print("The area of rectangle is: {}".format(self.perimeter()))
```

```python
R1= Rectangle(2,4) #object
R1.display()
```

```python
class BankAccount:
    
    def __init__(self, accountNumber, name, balance):
        self.accountNumber = accountNumber
        self.name = name
        self.balance = balance

    # Deposit Method
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print("Deposit successful.")
            print("Updated Balance:", self.balance)
        else:
            print("Invalid deposit amount.")

    # Withdrawal Method
    def withdrawal(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount.")
        elif amount > self.balance:
            print("Insufficient balance.")
        else:
            self.balance -= amount
            print("Withdrawal successful.")
            print("Remaining Balance:", self.balance)

    # Bank Fees Method (5%)
    def bankFees(self):
        fee = self.balance * 0.05
        self.balance -= fee
        print("Bank fees applied (5%).")
        print("Fee deducted:", fee)
        print("Updated Balance:", self.balance)

    # Display Method
    def display(self):
        print("Account Number:", self.accountNumber)
        print("Account Holder Name:", self.name)
        print("Account Balance:", self.balance)
```

```python
account1 = BankAccount(12345, "Sana", 1000)

account1.display()
account1.deposit(500)
account1.withdrawal(300)
account1.bankFees()
account1.display()
```

```python
# Q-3:Computation class
# Create a Computation class with a default constructor (without parameters) allowing to perform various calculations on integers numbers.
# Create a method called Factorial() which allows to calculate the factorial of an integer n. Integer n as parameter for this method
# Create a method called naturalSum() allowing to calculate the sum of the first n integers 1 + 2 + 3 + .. + n. Integer n as parameter for this method.
# Create a method called testPrime() in the Calculation class to test the primality of a given integer n, n is Prime or Not? Integer n as parameter for this method.
# Create a method called testPrims() allowing to test if two numbers are prime between them. Two integers are prime to one another if they have only 1 as their common divisor. Eg. 4 and 9 are prime to each other.
# Create a tableMult() method which creates and displays the multiplication table of a given integer. Then create an allTablesMult() method to display all the integer multiplication tables 1, 2, 3, ..., 9.
# Create a static listDiv() method that gets all the divisors of a given integer on new list called Ldiv. Create another listDivPrim() method that gets all the prime divisors of a given integer.
import math
class Computation():
    def __init__(self):
        pass

    def factorial(self, num):
        if num<0:
            factorial=math.factorial(num)
            print("The factorial of {} is: {}".format(self.num, factorial))
        else:
            print("Factorial not defined for negative numbers.")

    def natural_sum(self, num):
        if num < 0:
            print("Enter a positive integer.")
        else:
            total = num * (num + 1) // 2
            print(f"Sum of first {num} natural numbers is:", total)

    def testPrime(self, n):
        if n <= 1:
            print(n, "is NOT Prime")
            return False
        for i in range(2, int(math.sqrt(n)) + 1):
            if n % i == 0:
                print(n, "is NOT Prime")
                return False
        print(n, "is Prime")
        return True

    def tableMult(self, n):
        print(f"Multiplication Table of {n}")
        for i in range(1, 11):
            print(f"{n} x {i} = {n*i}")

    def allTablesMult(self):
        for n in range(1, 10):
            self.tableMult(n)
            print("-" * 20)

    def testPrims(self, a, b):
        if math.gcd(a, b) == 1:
            print(a, "and", b, "are prime to each other (Co-prime).")
        else:
            print(a, "and", b, "are NOT prime to each other.")

    @staticmethod
    def listDivisior(n):
        Ldiv = []
        for i in range(1, n+1):
            if n % i == 0:
                Ldiv.append(i)
        return Ldiv
    
    def listprimedivisiors(self, n):
        divisiors=self.listDivisior(n)
        prime_div=[]
        for d in divisiors:
            if self.testPrime(d):
                prime_div.append(d)
        return prime_div
```

```python
comp = Computation()

comp.factorial(5)
comp.natural_sum(10)
comp.testPrime(17)
comp.testPrims(4, 9)
comp.tableMult(3)
comp.allTablesMult()

print("Divisors of 12:", Computation.listDivisior(12))
print("Prime Divisors of 12:", comp.listprimedivisiors(12))
```

```python
import random

class FlashCard:

    def __init__(self):
        self.fruits = {
            "Banana": "yellow",
            "Strawberries": "pink",
            "Watermelon": "green",
            "Apple": "red",
            "Orange": "orange",
            "Grapes": "purple",
            "Mango": "yellow"
        }

    def play(self):   # <-- Method
        print("Welcome to Fruit Quiz!")

        while True:
            fruit, color = random.choice(list(self.fruits.items()))

            user_answer = input(f"What is the color of {fruit}? ").lower()

            if user_answer == color.lower():
                print("Correct answer ✅")
            else:
                print("Wrong answer ❌")
                print("Correct color is:", color)

            choice = input("Enter 0 if you want to play again: ")
            if choice != "0":
                break


# Create object
quiz = FlashCard()
quiz.play()
```

```python
class Instructor:
    def __init__(self):
        # private instance variables
        self.__name = ""
        self.__technology_skills = []   # list
        self.__experience = 0.0         # in years
        self.__avg_feedback = 0.0       # rating

    # -------- setters (public) --------
    def set_name(self, name):
        self.__name = name

    def set_technology_skills(self, skills_list):
        # expecting a list like ["Python", "AI", "Java"]
        self.__technology_skills = skills_list

    def set_experience(self, experience):
        self.__experience = experience

    def set_avg_feedback(self, avg_feedback):
        self.__avg_feedback = avg_feedback

    # -------- getters (public) --------
    def get_name(self):
        return self.__name

    def get_technology_skills(self):
        return self.__technology_skills

    def get_experience(self):
        return self.__experience

    def get_avg_feedback(self):
        return self.__avg_feedback

    # -------- required methods --------
    def check_eligibility(self):
        """
        Return True if instructor satisfies:
        - exp > 3  => feedback >= 4.5
        - exp <= 3 => feedback >= 4.0
        """
        if self.__experience > 3:
            return self.__avg_feedback >= 4.5
        else:
            return self.__avg_feedback >= 4.0

    def allocate_course(self, technology):
        """
        Return True if:
        - eligible by experience/feedback
        - has the required technology skill
        """
        if not self.check_eligibility():
            return False

        # skill match (case-insensitive)
        skills_lower = [s.lower() for s in self.__technology_skills]
        return technology.lower() in skills_lower


# ------------------ Testing ------------------
inst1 = Instructor()
inst1.set_name("Sana")
inst1.set_technology_skills(["Python", "ML", "SQL"])
inst1.set_experience(4)
inst1.set_avg_feedback(4.6)

print("Instructor:", inst1.get_name())
print("Eligible?:", inst1.check_eligibility())
print("Allocate Python course?:", inst1.allocate_course("Python"))
print("Allocate Java course?:", inst1.allocate_course("Java"))
print()

inst2 = Instructor()
inst2.set_name("Ali")
inst2.set_technology_skills(["Java", "Spring"])
inst2.set_experience(2)
inst2.set_avg_feedback(4.1)

print("Instructor:", inst2.get_name())
print("Eligible?:", inst2.check_eligibility())
print("Allocate Java course?:", inst2.allocate_course("Java"))
print("Allocate Python course?:", inst2.allocate_course("Python"))
print()

inst3 = Instructor()
inst3.set_name("Hira")
inst3.set_technology_skills(["Python", "Django"])
inst3.set_experience(3)
inst3.set_avg_feedback(3.9)  # not eligible (<=3 years needs >=4.0)

print("Instructor:", inst3.get_name())
print("Eligible?:", inst3.check_eligibility())
print("Allocate Python course?:", inst3.allocate_course("Python"))
```
