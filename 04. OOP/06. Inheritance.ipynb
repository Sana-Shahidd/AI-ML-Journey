1. Multiple Inheritance and Method Resolution Order (MRO)
Create the following class structure:
Class A with method show() printing "Class A".
Class B inherits from A and overrides show().
Class C also inherits from A and overrides show().
Class D inherits from both B and C.
Tasks:
Call show() using an object of D.
Print the MRO of class D.
Explain why that output occurs.


Class Structure

You created this inheritance hierarchy:
        A
       / \
      B   C
       \ /
        D

B inherits from A
C inherits from A
D inherits from B and C
This is called multiple inheritance.
When you call:
D1 = D()
D1.show()
Python first checks inside class D.
Since D already has show(), Python does not check B, C, or A.
So this runs:
Method overridden from inside class D.

```python
class A:
    def show(self):
        print("Class A.")
class B(A): 
    def show(self): #method overriding
        print("Class B.")
class C(A):
    def show(self):
        print("Class C.")
class D(B, C):
    def show(self):
        print("Method overridden from inside class D.")

D1=D()
D1.show()
```

2. Using super() in Multi-Level Inheritance

Create three classes:
Employee
Manager (inherits from Employee)
SeniorManager (inherits from Manager)
Requirements:
Employee should store name and salary.
Manager should add department.
SeniorManager should add bonus.
Tasks:
Use super() correctly in constructors.
Create a method total_compensation() in SeniorManager.
Print full details of the object.

```python
#Constructor overriding/method overriding
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department
    
class SeniorManager(Manager):
    def __init__(self, name, salary, department, bonus):
        super().__init__(name, salary, department )
        self.bonus= bonus
    
    def total_compensation(self):
        print(self.name, self.salary, self.department, self.bonus)

SM=SeniorManager("Ali", 80000, "IT", 20000)
SM.total_compensation()
```

1️⃣ First Rule: Always Check the MRO

Before reasoning about super(), always check the MRO.

print(Professor.mro())

Output:

[Professor, Teacher, Researcher, Person, object]

So Python will search in this order:

Professor → Teacher → Researcher → Person → object

Now whenever you see:

super()

ask yourself:

What is the next class in this order?
2️⃣ Second Rule: super() ≠ parent

Many beginners think:

super() → parent

But the real meaning is:

super() → next class in MRO

Example in your code:

Inside Teacher:

super().__init__(name)

Because the object is Professor, Python checks the Professor MRO.

After Teacher, the next class is:

Researcher

So Python behaves like:

Researcher.__init__(self, name)

Not:

Person.__init__(self, name)
3️⃣ Third Rule: Constructors must match arguments

If a constructor is defined like:

def __init__(self, name, subject, field)

then it must receive all arguments.

If Python calls it with fewer arguments → error.

How to avoid confusion in practice

When learning OOP, follow this safe pattern.

For simple learning

Avoid super() in multiple inheritance.

Use direct calls:

```python
class Person:
    def __init__(self, name):
        self.name = name

class Teacher(Person):
    def __init__(self, name, subject):
        Person.__init__(self, name)
        self.subject = subject

class Researcher(Person):
    def __init__(self, name, field):
        Person.__init__(self, name)
        self.field = field

class Professor(Teacher, Researcher):
    def __init__(self, name, subject, field):
        Teacher.__init__(self, name, subject)
        Researcher.__init__(self, name, field)

    def display_info(self):
        return self.name, self.subject, self.field

P1 = Professor("sana", "Gen-AI", "Artificial Engineering")
print(P1.display_info())
```

4. Abstract Base Class with Inheritance

Create an abstract class Vehicle.

Requirements:
Abstract methods:
start_engine()
stop_engine()
Create two subclasses:
Car
Motorcycle
Tasks:

Implement both abstract methods.
Add a method vehicle_type().
Create a list of mixed vehicle objects and demonstrate polymorphism.
return super().stop_engine()??

```python
from abc import ABC, abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        raise NotImplementedError("Method does not belong to class vehicle")
    
    @abstractmethod
    def stop_engine(self):
        raise NotImplementedError("Method does not belong to class vehicle")

class Car(Vehicle):
    def start_engine(self):
        print("car started...")
    def stop_engine(self):
        print("Car stopped....")
    def vehicle_type(self):
        print("Car")

class Motorcycle(Vehicle):
    def start_engine(self):
        print("motorcycle started...")
    def stop_engine(self):
        print("motorcycle stopped....")
    def vehicle_type(self):
        print("Motorcycle")

# Polymorphism
vehicles = [Car(), Motorcycle()]
for v in vehicles:
    v.vehicle_type()
    v.start_engine()
    v.stop_engine()
    print()
```

Static methods overriding

Static methods can be overridden (redefined) in subclasses, but they do not support true polymorphic overriding because they are not bound to instances or classes.

```python
class A:
    @staticmethod
    def show():
        print("A")

class B(A):
    @staticmethod
    def show():
        print("B")

a = A()
b = B()

a.show()
b.show()

# Static methods do not receive self or cls
# Because they are not bound to an object or class instance.
# They behave like normal functions placed inside a class namespace.
# In subclasses, a static method can be overridden (redefined), but it does not participate in polymorphism like instance methods.
```

5. Method Overriding with Parent Method Call

Create a base class Shape.

Requirements:
Method area()
Method describe() that prints "This is a shape".

Create subclasses:
Rectangle
Circle

Tasks:
Override describe() in both subclasses.
Still call the parent describe() using super().
Add correct area calculations.

```python
import math
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def Area(self):
        pass

    def describe(self):
        print("This is a shape")

class Rectangle(Shape):
    def __init__(self, length, width):
        self.l=length
        self.w=width
    def Area(self):
        super().describe()
        return self.l*self.w
    
    def __str__(self):
        return f"Area = {self.Area()}"
    
    def describe(self):
        print("This is rectangle..")

class Circle(Shape):
    def __init__(self, radius):
        self.r=radius
        
    def Area(self):
        super().describe()
        return self.r*math.pi
    
    def __str__(self):
        return f"Area = {self.Area()}"
    
    def describe(self):
        print("This is circle..")

shapes=list([Rectangle(2,4),Circle(2)])
for s in shapes:
    s.describe()
    print(s)
    print()
```

Key Concepts to Cover in Inheritance

Inheritance is a fundamental concept of Object-Oriented Programming that allows a class to reuse and extend the behavior of another class. When learning inheritance in Python, the following concepts should be clearly understood.

1. Types of Inheritance

Understand the different inheritance structures used in Python:

Single Inheritance – one child inherits from one parent

Multilevel Inheritance – inheritance chain across multiple levels

Hierarchical Inheritance – multiple child classes inherit from one parent

Multiple Inheritance – one child inherits from multiple parents

Hybrid/Diamond Inheritance – combination of multiple inheritance structures

2. Method Overriding

Child classes can redefine methods of the parent class to provide specialized behavior. This enables runtime polymorphism.

3. Use of super()

super() is used to call methods or constructors of the parent class. In multiple inheritance, super() follows the Method Resolution Order (MRO) rather than the direct parent.

4. Method Resolution Order (MRO)

Python determines the order in which classes are searched for methods using C3 Linearization. Understanding MRO is crucial in multiple inheritance.

5. Constructor Inheritance

Child classes can:

inherit constructors from the parent

extend them

or override them using super().

6. Abstract Base Classes

Using ABC and @abstractmethod, parent classes can enforce implementation of specific methods in subclasses.

7. Polymorphism through Inheritance

Objects of different subclasses can be treated uniformly when they share a common parent interface.

8. Overriding vs Method Replacement

Understand differences between:

instance method overriding

class method overriding

static method redefinition

9. Code Reusability and Extensibility

Inheritance promotes code reuse, modular design, and easier maintenance by extending existing functionality rather than rewriting code.

```python
import math
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def Area(self):
        pass

    @abstractmethod
    def Perimeter(self):
        pass
    
    #normal method
    def describe(self):
        print("This is a shape")

    @staticmethod
    def shape_category():
        print("Shapes belong to geometry")
    
class Rectangle(Shape):
    def __init__(self, length, width):
        self.l = length
        self.w = width

    # method overriding
    def Area(self):
        return self.l* self.w

    def Perimeter(self):
        return 2*(self.l+self.w)
    
    # Override describe() and call the parent method using super()
    def describe(self):
        super().describe()
        print("This shape is rectangle")

    # Redefine the static method shape_category()
    @staticmethod
    def shape_category(): 
        print("Rectangle belong to geometry")
 
    # magic method for display
    def __str__(self):
        return f"Area of rectangle is: {self.Area()} \n Perimeter of rectangle is: {self.Perimeter()}"

class Circle(Shape):
    def __init__(self, radius):
        self.r = radius

    # method overriding
    def Area(self):
        return math.pi*self.r**2

    def Perimeter(self):
        return 2*math.pi*self.r
    
    # Override describe() and call the parent method using super()
    def describe(self):
        super().describe()
        print("This shape is circle")

    # Redefine the static method shape_category()
    @staticmethod
    def shape_category(): 
        print("Circle belong to geometry")

    # magic method for display
    def __str__(self):
        return f"Area of Circle is: {self.Area()} \n Perimeter of Circle is: {self.Perimeter()}"

class ColoredRectangle(Rectangle):
    def __init__(self,length,width,colour):
        super().__init__(length, width)
        self.colour=colour
    
    #method override
    def describe(self):
        print("Colour of rectangle is : ", self.colour)

shapes=[Circle(radius=1),Rectangle(length=2, width=3),ColoredRectangle(length=4, width=5, colour="Red")]
for s in shapes:
    s.describe()
    s.Perimeter()
    s.Area()

Shape.shape_category()
Rectangle.shape_category()
Circle.shape_category()
ColoredRectangle.mro()
```
