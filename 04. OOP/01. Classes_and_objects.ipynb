Two tpes of classes in python:

```python
class ATM:
    
    def __init__(self): #__funcName__ is magic functions
        self.pin =''
        self.balance = 0
        self.is_created=False
        print(id(self))
        self.menu()

    def menu(self):
        user_input = input('''Welcome to AN Bank. How can i help you!
        1 for creating the pin
        2 for changing the pin
        3 for checking the balance
        4 for withdrawal of amount
        5.anything else to exit
        ''')
        if user_input == '1':
            self.create_pin()
        elif user_input=='2':
            self.change_pin()
        elif user_input =='3':
            self.check_balance()
        elif user_input == '4':
            self.withdraw()
        else:
            exit
    def create_pin(self):
        user_pin=input("Enter ypur pin")
        self.pin= user_pin
        self.balance=int(input("Enter your balance amount"))
        self.is_created=True
        print("Pin created successfully")
        self.menu()

    def change_pin(self):
        if self.is_created == True:
            old_pin=input("Enter your old pin: ")
            if old_pin == self.pin:
                new_pin=input("Enter the new pin")
                confirm_pin=input("Enter the pin again")
                if new_pin==confirm_pin:
                    print("Pin changed successfully")
                else:
                    print("Wrong pin")
                    print("Confirm the pin again")
                    confirm_pin=input("Enter the pin again")
        else:
            print("create the pin first")
        self.menu()
    
    def check_balance(self):
        if self.is_created == True:
            user_pin=input("Enter the pin first: ")
            if user_pin == self.pin:
                print("The balance of your account is {}".format(self.balance))
            else:
                print("Wro5" \
                "ng pin.Enter your pin again")
                user_pin=input("Enter the pin first: ")
        else:
            print("create the account first")
        self.menu()
    
    def withdraw(self):
        if self.is_created == True:
            user_pin = input("Enter your pin: ")
            if user_pin == self.pin:
                amount = int(input("Enter the amount to withdraw: "))
                if amount <= 0:
                    print("Enter a valid amount")
                elif amount <= self.balance:
                    self.balance -= amount
                    print(f"Withdrawal successful. Remaining balance: {self.balance}")
                else:
                    print("Insufficient balance")
            else:
                print("Wrong pin")
        else:
            print("Create the account first")
        self.menu()
```
