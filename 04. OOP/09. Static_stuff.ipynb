```python
# Can static methods access class variables?
# Not directly through cls, because cls is not available.
# But they can access class variables using the class name.
class Demo:
    x = 100

    @staticmethod
    def show():
        print(Demo.x)

Demo.show()
```

```python
# Can static methods access instance variables?
# Not directly.
# Example:

class Demo:
    def __init__(self, x):
        self.x = x

    @staticmethod
    def show():
        print(self.x)   # Error

# This gives error because self does not exist inside static method.
# If you really want instance data, then make it an instance method, not static.
```

```python
# Static methods and inheritance
# Static methods are inherited by child classes.
# Example:

class Parent:
    @staticmethod
    def greet():
        print("Hello from Parent")

class Child(Parent):
    pass

Child.greet()

# Output:
# Hello from Parent
# So child classes can use static methods too.
```

```python
# Real-life analogy

# Suppose you have a class Employee.
# instance method: shows one employee’s salary
# class method: changes company name for all employees
# static method: checks whether an age is valid for hiring

# Example:

class Employee:
    company = "TechSoft"

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show(self):
        print(self.name, self.salary)

    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name

    @staticmethod
    def is_valid_age(age):
        return 18 <= age <= 60

# Here is_valid_age() does not depend on one employee or the company itself. It is just a related helper.
```
