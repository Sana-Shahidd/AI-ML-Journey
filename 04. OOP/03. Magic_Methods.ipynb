```python
class Fractions:
    def __init__(self, x, y):
        self.num= x
        self.denum= y

    def __str__(self): #mera object dekh kesa rha ha print ke andar star ka code dalo gae to print ho jai ga
        return "{}/{}".format(self.num,self.denum)
    
    def __add__(self, other):
        new_num = self.num * other.denum + other.num * self.denum
        new_den = self.denum * other.denum
        return "{}/{}".format(new_num, new_den)
    
    def __sub__(self, other):
        new_num = self.num * other.denum - other.num * self.denum
        new_den = self.denum * other.denum
        return "{}/{}".format(new_num, new_den)
    
    def __mul__(self, other):
        new_num = self.num * other.num
        new_den = self.denum * other.denum
        return "{}/{}".format(new_num, new_den)
    
    def __truediv__(self, other):
        new_num = self.num * other.denum
        new_den = self.denum * other.num
        return "{}/{}".format(new_num, new_den)
```

```python
fr1=Fractions(3,4)
print(fr1)
fr2=Fractions(4,5)
print(fr1+fr2)
print(fr1-fr2)
print(fr1 * fr2)
print(fr1 / fr2)
```
