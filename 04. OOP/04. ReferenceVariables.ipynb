```python
# A user can create and view 2D coordinates
class Points:
    def __init__(self,x,y):
        self.cord_x=x
        self.cord_y=y
    def __str__(self):
        return "<{},{}>".format(self.cord_x,self.cord_y)
#A user can find out the distance between 2 coordinates
    def euclideans_distance(self, other):
        return ((self.cord_x-other.cord_x)**2+(self.cord_y-other.cord_y)**2)**0.5
# Ditance from the center
    def distance_from_origin(self):
        return self.euclideans_distance(Points(0,0))
#A user check if a given point lies on line.
class Line:
    def __init__(self,A,B,C):
        self.A= A
        self.B= B
        self.C= C

    def __str__(self):
        return "{}x + {}y +{}=0>".format(self.A,self.B,self.C)  
    
    def point_on_line(line,points):
        if line.A*points.cord_x+line.B*points.cord_y+line.C==0:
            return "lies on line"
        else:
            return "doesnot lies on line"
        
    def shortest_distance(line, point):
        return abs(line.A*point.cord_x + line.B*point.cord_y+line.C)/((line.A)**2+(line.B)**2)**0.5
    
    def check_lines_intersection(line_1, line_2):
        '''
        I am using the standard line equation Ax+By+C 
        if determinent is non zero that means lines intersect if determinent is zero you have to check furthur two things if A1/A2==B1/B2==C1/C2 
        lines coincidates else line is parallel and for coordinates x=(B1C2*B2C1)/(A1B2*A2B1) for y=(C1A2*C2A1)/(A1B2*A2B1)
        '''
        determinent=(line_1.A*line_2.B)-(line_2.A*line_1.B)
        if determinent != 0:
            x_intersect_coordinate=(line_1.B*line_2.C - line_2.B*line_1.C)/(line_1.A*line_2.B - line_2.A*line_1.B)
            y_intersect_coordinate=(line_1.C*line_2.A - line_2.C*line_1.A)/(line_1.A*line_2.B - line_2.A*line_1.B)
            print("The x and y coordinates are {},{}".format(x_intersect_coordinate,y_intersect_coordinate))
            print("Yes the lines intersect each other")
        elif determinent==0:
            if (line_1.A/line_2.A == line_1.B/line_2.B == line_1.C/line_2.C):
                x_intersect_coordinate=(line_1.B*line_2.C - line_2.B*line_1.C)/(line_1.A*line_2.B - line_2.A*line_1.B)
                y_intersect_coordinate=(line_1.C*line_2.A - line_2.C*line_1.A)/(line_1.A*line_2.B - line_2.A*line_1.B)
                print("The x and y coordinates are {},{}".format(x_intersect_coordinate,y_intersect_coordinate))
                print("The lines are coincident and they intersect with each other")
            else:
                print("The lines are parallel and they donot intersect with each other.")

    
```
