```python
# Agregation(weak has a)
# Objects can exist independently.
class Engine:
    pass

class Car:
    def __init__(self, engine):
        self.engine = engine

# Here:
# Engine exists separately
# Car just uses the engine
# So this is aggregation.
```

```python
# Composition(strong has a relation)
# object cannot exit independently
class Engine:
    pass

class Car:
    def __init__(self):
        self.engine = Engine()

# Here:
# Engine is created inside Car
# Engine cannot exist without Car
# So this is composition.
```
