```python
#first thought revision
#list flattening
lst=[[1,2, [3,4,[9,0,4]]]]
#first way
def flatten(lst):
    result=[]
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result
print(flatten(lst))
#second way (list comprehension works till one level)
lst = [[1,2], [3,4]]
result=[ele for item in lst for ele in item]
print(result)
```

```python
# 1.	Invert dictionary { "a":1, "b":2 } → {1:"a", 2:"b"}.
d3={ "a":1, "b":2 }
d4={v:k for k,v in d3.items()}
print(d4)
# 2.	Sort dictionary by values (ascending).
d3={ "a":2, "b":1 }
r_ascending=sorted(d3.items(), key=lambda x: x[1])
print(r_ascending)
# 3.	Sort dictionary by keys (descending).
d3={ "a":1, "b":2 }
r_descending=sorted(d3.items(), key=lambda x: x[1], reverse=True)
print(r_descending)
# 4.	Merge two dictionaries and sum values if key exists in both.
d3={ "a":2, "b":1 }
d4={ "a":1, "b":2 }
d5=d3.copy()
for key, val in d4.items():
    d5[key]=d5.get(key, 0)+ val #existing value + new value
print(d5)
# 5.	Remove duplicate values from dictionary.
d6 = {"a":1, "b":2, "c":2}
seen = set()
result = {}
for key, val in d6.items():
    if val not in seen:
        result[key] = val
        seen.add(val)
print(result)
#second way
d6 = {"a":1, "b":2, "c":2}
d7 = d6.copy()
for key1, val1 in d6.items():
    count = 0
    for key2, val2 in d7.items():
        if val1 == val2:
            count += 1
    if count > 1:
        d7.pop(key1)
print(d7)

# 6.	Count the no of items of characters in a string.
d6 = {"a":"apple", "b":"banana", "c":"mango"}
key=list(d6.keys())
items=list(d6.values())
count=[]
for word in items:
    count.append(len(word))
print(count)
words_count={k:v for k, v in zip(key,count)}
print(words_count)
# Count frequency of characters in a string.
from collections import Counter
d6 = {"a":"apple", "b":"banana", "c":"mango"}
count={}
for word in d6.values():
    count.update(Counter(word))
print(count)
# 7.	Group words by their length.
d6 = {"a":"apple", "b":"banana", "c":"mango"}
grouped = {}
for word in d6.values():
    length = len(word)
    if length not in grouped:
        grouped[length] = []
    grouped[length].append(word)
print(grouped)
# 8.	Group numbers by even/odd:
# o	{ "even":[...], "odd":[...] }
d6 = {"a":1, "b":2, "c":3}
grouped={"even": [], "odd": []}
for key,val in d6.items():
    if val%2==0:
        grouped["even"].append(val)
    else:
        grouped["odd"].append(val)
print(grouped)

# 8a. words = ["apple", "banana", "apricot", "blueberry", "cherry"]
#     Group them by their first letter.

words = ["apple", "banana", "apricot", "blueberry", "cherry"]
grouped={}
for word in words:
    initial=word[0]
    if initial not in grouped:
        grouped[initial]=[]
    grouped[initial].append(word)
print(grouped)

# 8b. Group Numbers by Range (Bucket Grouping)
# nums = [5, 12, 7, 25, 18, 3, 30]
# Group them into ranges of 10:
# 0–9
# 10–19
# 20–29
# 30–39

nums = [5, 12, 7, 25, 18, 3, 30]
grouped={}
for item in nums:
    lowerlimit= (item//10)*10
    upperlimit = lowerlimit + 9
    key = f"{lowerlimit}-{upperlimit}"
    if key not in grouped:
        grouped[key] = []
    grouped[key].append(item)
print(grouped)


# Given:
# words = ["eat", "tea", "tan", "ate", "nat", "bat"]
# Group anagrams together.
# Expected Output:
# [
#   ["eat", "tea", "ate"],
#   ["tan", "nat"],
#   ["bat"]
# ]

words = ["eat", "tea", "tan", "ate", "nat", "bat"]
grouped = {}
for word in words:
    key = tuple(sorted(word))   # tuple because lists can't be dict keys
    if key not in grouped:
        grouped[key] = []
    grouped[key].append(word)
print(list(grouped.values()))

# 9.	Create nested dictionary for students:
# o	{ "Ali": {"math":90, "eng":85} }

nested_dic={ "Ali": {"math":90, "eng":85} }
print(nested_dic)

# 10.	Update nested dictionary value.
nested_dic["Ali"]["physice"]=90
print(nested_dic)

# 11.	Flatten nested dictionary one level.
flattened = {}

for key, value in nested_dic.items():
    if isinstance(value, dict):
        for sub_key, sub_value in value.items():
            flattened[f"{key}_{sub_key}"] = sub_value
    else:
        flattened[key] = value
print(flattened)
```
