```python
# Problem 1: Combine two lists index-wise (column-wise)
list1 = ["M", "na", "i", "Kh"]
list2 = ["y", "me", "s", "an"]

result = [[a, b] for a, b in zip(list1, list2)]
print(result)
```

```python
# Problem 2: Add 7000 after 6000
list1 = [10, 20, [300, 400, [5000, 6000], 500], 30, 40]

list1[2][2].insert(2, 7000)
print(list1)
```

```python
# Problem 3: Show candy items with counts
candy_list = ['Jelly Belly','Kit Kat','Double Bubble','Milky Way','Three Musketeers']
no_of_items = [10,20,34,74,32]

for candy, qty in zip(candy_list, no_of_items):
    print(f"{candy}-{qty}")
```

```python
# Problem 4: Running sum
list1 = [1,2,3,4,5,6]

running = []
total = 0
for x in list1:
    total += x
    running.append(total)

print(running)
```

```python
# Problem 5: For each element, sum(all elements >= it) (as per your example)
lst = [2,4,6,10,1]

result = []
for x in lst:
    s = sum(v for v in lst if v >= x)
    result.append(s)

print(result)   # [22, 20, 16, 10, 23]
```

```python
# Problem 6: Find list of common unique items from two list. and show in increasing order
# Input
# num1 = [23,45,67,78,89,34]
# num2 = [34,89,55,56,39,67]
# Output:
# [34, 67, 89]
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
unique=[x for x in num1 if x in num2  ]
print(unique)
```

```python
lst = ['CampusX is a channel', 'for data-science', 'aspirants.']
result = [word for sentence in lst for word in sentence.split()]
print(result)
```

```python
words = ['campusxIs', 'bestFor', 'dataScientist']

result = ["".join((" " + ch) if ch.isupper() else ch for ch in w).strip() for w in words]
print(result)
```

```python
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

transposed = [[matrix[r][c] for r in range(len(matrix))] for c in range(len(matrix[0]))]
print(transposed)
```
