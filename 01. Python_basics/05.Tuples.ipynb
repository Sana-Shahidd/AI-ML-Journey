```python
# converting 2D list to 1D list
t = (1, (2, 3), (4, (5, 6)))
lst=list(t)
def flattening(lst):
    result=[]
    for item in lst:
        if isinstance(item, (list, tuple)):
            result.extend(flattening(item))
        else:
            result.append(item)
    return result

ret=flattening(lst)
t=tuple(ret)
new_t=tuple(x for x in t if x % 2 == 0)
print(new_t)
```

Why Deduplication Works
Key Concept:

Sets only store unique hashable elements.

Tuples are:

✔ Immutable

✔ Hashable (if all elements inside are hashable)

So Python can store them inside a set.

```python
# Create a set of tuples from list [(1,2), (1,2), (2,3)] and explain deduplication.
l=[(1,2), (1,2), (2,3)]
a=set(l)
a
```

```python
# Given students = (("A", 20, 90), ("B", 20, 85), ("C", 19, 90)), sort by score desc, age asc, name asc.
students = (("A", 20, 90),
            ("B", 20, 85),
            ("C", 19, 90))

result = tuple(sorted(students, key=lambda x: (-x[2], x[1], x[0])))

print(result)
```
