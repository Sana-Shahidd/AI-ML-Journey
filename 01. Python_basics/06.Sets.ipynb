Adding the elments(*add for singe item *update for multiple items)

```python
s={1,2,3,4}
s.add(5)
print(s)
s.update({6,7,8,9})
print(s)
```

Delete(*del *discard *remove *pop *clear)

```python
s={1,2,3,4}
del(s)
s={1,2,3,4,5}
s.remove(3) #if ele not present this throws errror
print(s)
s.discard(5) #discart does not thrws error if ele no present
print(s)
s.pop() #randomly deletes an element
print(s)
s.clear()
print(s)
```

-operations{union(|), intersection(&) difference(-), symmetric difference(^), membership test, iterations}

```python
set1={1,2,3}
set2={2,4,5}
set1 | set2
set1 & set2
set1 - set2
set1 ^ set2 #(prints all the otherwise elements of both sets)
1 in set1
[i*j for i in set1 for j in set2] # all 3 comprehensions
{i*j for i in set1 for j in set2}
tuple(i+j for i,j in zip(set1, set2))
```
