Shallow Copy (independent copy)

a) Using slicing

```python
List1 = [1, 2, 3, 4]
List2=List1[:]
print(id(List1))
print(id(List2))
List1.append(5)
print(List1)  #list 1 ke change krne se list 2 me change nhi ata 
print(List2)
```

```python
# b) Using list() constructor
List2 = list(List1)
print(List1)  #list 1 ke change krne se list 2 me change nhi ata 
print(List2)
print(id(List1))
print(id(List2))
```

Deep copy (for nested lists)

If the list has nested lists, shallow copy only copies outer list. Use copy.deepcopy for full independent copy:

```python
import copy

list1 = [1, [2, 3], 4]
list2 = copy.deepcopy(list1)

list2[1].append(5)
print(list1)  # [1, [2, 3], 4]
print(list2)  # [1, [2, 3, 5], 4]
```

Write a program to count how many times an element appears in a list.

```python
List3=[1,2,2,3,3,4,5,5,5,5,6,6,7,8,9]
List3.count(3)
freq_dict = {}
for item in set(List3):
    freq_dict[item]=List3.count(item)

print(freq_dict)
```

```python
from collections import Counter
List3=[1,2,2,3,3,4,5,5,5,5,6,6,7,8,9]
freq_dict = Counter(List3)
freq_dict
```
