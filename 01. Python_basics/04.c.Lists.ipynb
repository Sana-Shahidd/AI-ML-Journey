```python
#shallow copy
a=[1,2,3]
b = a[:]
print(id(a))
print(id(b))

a.append(5)

print(a)
print(b)
```

```python
import copy

a = [1, 2, 3]
b = copy.copy(a)
print(a)
print(b)
a.append(5)
print(a)
print(b)
```

```python
# Shallow copy only copies the outer container. inner lists are shared 
a=[[1,2,3],[4,5]]
b=a.copy()
print(a)
print(b)
a.append(5)
a[0].append(9)
print(a)
print(b)
```

```python
import copy

a = [[1, 2], [3, 4]]
b = copy.deepcopy(a)

print(id(a))
print(id(b))

a[0].append(99)

print(a)
print(b)
```

```python
| Type         | Outer ID  | Inner ID  | Change A affects B? |
| ------------ | --------- | --------- | ------------------- |
| Assignment   | Same      | Same      | Yes                 |
| Shallow Copy | Different | Same      | Sometimes           |
| Deep Copy    | Different | Different | No                  |
```
