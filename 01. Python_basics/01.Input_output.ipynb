Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

```python
x=int(input("Enter the first number: "))
y=int(input("Enter the second number: "))
eucledians_distance=((y-x)**2 + (y-x)**2)**0.5
print(f'The eucledians distance for two points ({x},{y}) is: {eucledians_distance}')
```

Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.

```python
principle_value=float(input("Enter the principle value: "))
interest_rate=float(input("Enter the interest rate: "))
Time_period=float(input("Enter the time period: "))
simple_interest=(principle_value*interest_rate*Time_period)/100
print(f'The simple interest for with principle_value of {principle_value}, interest_rate of {interest_rate} and {Time_period} is: {simple_interest}')
```

Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.
For example: Input: heads -> 4 legs -> 12
Output: dogs -> 2 chicken -> 2

```python
heads=int(input("Enter the heads: "))
legs=int(input("Enter the legs: "))
# d+c=heads -> c=heads-((legs- 2*heads)/2)
# 4d + 2c = legs ->4d + 2(heads-d)=legs-> 2d + 2*heads= legs -> d= (legs- 2*heads)/2
chickens=heads-((legs- 2*heads)//2)
dogs=(legs- 2*heads)//2
print(f"The number of chikens and dogs are {chickens},{dogs} respectively")
```

Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

```python
a = int(input("Enter first term: "))
b = int(input("Enter second term: "))
n = int(input("Enter the value of n: "))

# Calculate common difference
d = b - a

# Calculate nth term
nth_term = a + (n - 1) * d

print("The", n, "th term of the series is:", nth_term)
```

Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.
Input:
Dimensions of the milk tank
H = 20cm, L = 20cm, B = 20cm

Dimensions of the glass
h = 3cm, r = 1cm

```python
import math

# Tank dimensions
H = float(input("Enter tank height (cm): "))
L = float(input("Enter tank length (cm): "))
B = float(input("Enter tank breadth (cm): "))

# Glass dimensions
h = float(input("Enter glass height (cm): "))
r = float(input("Enter glass radius (cm): "))

# Volume calculations
tank_volume = H * L * B
glass_volume = math.pi * r * r * h

# Number of full glasses
number_of_glasses = int(tank_volume // glass_volume)

print("Number of glasses that can be filled:", number_of_glasses)
```
