```python
#copy
# 1. shallow copy (inner ids same, outer differnt)
#  1a.uisng slicing
a=[1,2,3]
# b=a #this does not cretes original copy you can prove this using oid function
b=a[:]
print(a)
print("id of a: " , id(a))
print(b)
print("id of b: " , id(b))
#  1b.using inbuilt method.copy
a=[1,2,3]
b=a.copy()
print(a)
print("id of a: " , id(a))
print(b)
print("id of b: " , id(b))
#  1c.using list constructor 
a = [1, 2, 3]
b = list(a)
print(a)
print("id of a: " , id(a))
print(b)
print("id of b: " , id(b))
# 2. deep copy (inner ids different, outer differnt)
#  2b.using inbuilt method.deepcopy
import copy
a=[1,2,3]
b=copy.deepcopy(a)
print(a)
print("id of a: " , id(a))
print(b)
print("id of b: " , id(b))
```

use shallow copy when data inside the list is immutable and and it is fast and safe
use deep copy when data inside the list is mutable(list, dict) and and it is fast and safe, JSON-like data
Ask yourself:

"If I modify a nested object, should the copy change?"

If YES → Shallow is fine(blue print)
If NO → Use Deep copy(house)

```python
#flatening the multi dimensional list

def flatten(lst):
    result=[]
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result
lst=[1,2,[3,4],[5,6,7,8,[9,0,8,7], 5, 6], 7,8,6,9,0]
flatten(lst)
```
