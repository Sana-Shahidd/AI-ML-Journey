```python
# Problem 3: Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.
# Sample String : 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
# Expected Output :
# No. of Upper case characters :  9
# No. of Lower case Characters :  47
def calc_upper_lower(sentence):
    count_upper = 0
    count_lower = 0
    
    for i in sentence:
        if i.isupper():
            count_upper += 1
        elif i.islower():
            count_lower += 1
    
    return count_upper, count_lower
sentence='CampusX is an Online Mentorship Program fOr EnginEering studentS.'
ans=calc_upper_lower(sentence)
print("The upper count is {} and lower count is {}".format(ans[0],ans[1]))
```

```python
# Problem 4: Write a Python program to print the even numbers from a given list.
# Sample List : [1, 2, 3, 4, 5, 6, 7, 8, 9]
# Expected Result : [2, 4, 6, 8]
def is_even(lst):
    result=[]
    for i in lst:
        if i%2==0:
            result.append(i)
    return result
lst=[1, 2, 3, 4, 5, 6, 7, 8, 9]
is_even(lst)          
```

```python
# Problem-6: Write a Python function to concatenate any no of dictionaries to create a new one.
# Sample Dictionary :
# dic1={1:10, 2:20}
# dic2={3:30, 4:40}
# dic3={5:50,6:60}
# Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}
def dict_sum(*args):
    result = {}
    for d in args:
        result.update(d)
    return result

dict1 = {1:10, 2:20}
dict2 = {3:30, 4:40}
dict3 = {5:50, 6:60}

print(dict_sum(dict1, dict2, dict3))
#secondway
result={**dict1, **dict2, ** dict3}
print(result)
#third way
def dict_sum(*args):
    return {k:v for d in args for k,v in d.items()}
```

```python
# Problem-7 Write a python function that accepts a string as input and returns the word with most occurence.
# Input:
# hello how are you i am fine thank you
# Output
# you -> 2

from collections import Counter
def count_occurance(sentence):
    words=sentence.split()
    count=Counter(words)
    word, freq=count.most_common(1)[0] #(1) no of return values [0] to get tuple itself
    return word, freq
sentence='hello how are you i am fine thank you'
count_occurance(sentence)
```

```python
def count_occurance(sentence):
    words = sentence.split()
    count = Counter(words)
    word = max(count, key=count.get)
    return word, count[word]
```

```python
# Problem-8 Write a python function that receives a list of integers and prints out a histogram of bin size 10
# Input:
# [13,42,15,37,22,39,41,50]
# Output:
# {11-20:2,21-30:1,31-40:2,41-50:3}
def histogram(lst):
    bins = {}
    
    for num in lst:
        lower = (num // 10) * 10 + 1
        upper = lower + 9
        key = f"{lower}-{upper}"
        bins[key] = bins.get(key, 0) + 1
    
    return dict(sorted(bins.items()))
lst=[13,42,15,37,22,39,41,50]
histogram(lst)
```

```python
# Problem-9 Write a python function that accepts a list of 2D co-ordinates and a query point, and then finds the the co-ordinate which is closest in terms of distance from the query point.
# List of Coordinates
# [(1,1),(2,2),(3,3),(4,4)]
# Query Point
# (0,0)
# Output
# Nearest to (0,0) is (1,1)
def nearest_point(points, query):
    return min(points, key=lambda p: (p[0] - query[0])**2 + (p[1] - query[1])**2)

points = [(1,1),(2,2),(3,3),(4,4)]
query = (0,0)

result = nearest_point(points, query)
print(f"Nearest to {query} is {result}")
```

```python
# Problem-12:Write a Python program to create a list containing the power of said number in bases raised to the corresponding number in the index using Python map.
# Input:
# list1 = [1,2,3,4,5,6]
# Output:
# [1,2,9,64,625,-]
list1 = [1,2,3,4,5,6]
ans = list(map(lambda x: x[1] ** x[0], enumerate(list1)))
print(ans)
```

```python
def filter_vowels(text):
    vowels = "aeiou"
    return list(filter(lambda ch: ch.lower() in vowels, text))

s = "Hello World"
print(filter_vowels(s))
```
