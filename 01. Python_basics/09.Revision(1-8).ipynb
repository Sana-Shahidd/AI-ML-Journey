Problem 1: Combine two lists index-wise(columns wise)
Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list, then the 1st index item, and so on till the last element. any leftover items will get added at the end of the new list.

Given List:

list1 = ["M", "na", "i", "Kh"]
list2 = ["y", "me", "s", "an"]
Output:

[['M','y'], ['na', me'], ['i', 's'], ['Kh', 'an']]

```python
list1 = ["M", "na", "i", "Kh"]
list2 = ["y", "me", "s", "an"]
list3=[(x,y) for x,y in zip(list1, list2)]
print(list3)
```

```python
dic={
    'name':'sana',
    'age':20
}
print(dic)
print(dic.keys())
print(dic.values())
print(dic.items())
print(dic.get('name'))
#alternative of get
print(dic['name'])
dic['sem']='5th'
print(dic)
# Count frequency of items in a list using dict.
lst=[1,2,3,2,2,3,5]
frequency={}
for item in lst:
    frequency[item]=frequency.get(item,0)+1 #agr frequency ke andar item nhi ha to uska number by default 0 set kr do wrna har dafa +1 kr do
print(frequency)
# Remove a key from dict without raising error if missing.
d = {"a": 1}
print(d.pop('a',None)) #pop kr do a ko agr nhi ha to none return kr do
# Invert dict: {"a":1,"b":2} → {1:"a",2:"b"}
dic={"a":1,"b":2}
inverted={value:key for key, value in dic.items()}
print(inverted)
# Merge two dicts with summing values on collisions.
dic1 = {"a":1,"b":2}
dic2 = {"c":3,"d":4}
merged = dic1.copy()
for key, value in dic2.items():
    merged[key] = merged.get(key, 0) + value
print(merged)
#Why This Pattern Is Powerful
# This pattern is used for:
# Frequency counting
# Merging data
# Accumulating totals
# Combining reports
# Aggregating values
```

```python
# Group words by length: ["hi","cat","lion"] → {2:["hi"],3:["cat"],4:["lion"]}
lst=["hi","cat","lion"]
grouped = {}
for word in lst:
    length = len(word)
    grouped[length] = grouped.get(length, []) + [word]
print(grouped) 
```
