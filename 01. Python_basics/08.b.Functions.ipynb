A lambda is just a small, anonymous (unnamed) function that is usually:
Short
Written in one line
Used temporarily
Passed as an argument to another function

```python
# 1. Return Last Character Safely
# Task: Return last character of a string. If empty → return ""
# Input:
# ["hello", "A", ""]
# Output:
# ["o", "A", ""]

lst=["hello", "A", ""]
last_characters=list(map((lambda x: x[-1] if x else ""), lst ))
last_characters


```

```python
# 2. Sort Strings by Length then Alphabetically
# Input:
# ["banana", "pie", "apple", "kiwi", "bat"]
# Output:
# ["bat", "pie", "kiwi", "apple", "banana"]
# (Shorter strings first; same length → alphabetical)
fruits=["banana", "pie", "apple", "kiwi", "bat"]
ans=list(sorted(fruits, key=lambda x:(len(x),x) ))
ans
```

```python
# 3. Sort Numbers Using (is_even, n)
# Input:
# [5, 2, 8, 1, 3, 4]
# Output:
# [2, 4, 8, 1, 3, 5]
# (Even numbers first, then odd — each group sorted naturally)

lst= [5, 2, 8, 1, 3, 4]
ans=list(sorted(lst, key=lambda x: ((x%2,x))))
ans
```

```python
# 4. Correctly Capture Loop Variable in Lambdas
# Input:
# Creating 3 lambdas in loop (0,1,2)
# Calling each with 10
# Output:
# [10, 11, 12]
# (Each lambda remembers its own loop value)

func=[]
for i in range(3):
    func.append(lambda x, i=i: x+i)
print([f(10) for f in func])
```

5. Convert List of Tuples → List of Dicts
Input:
[("Ali", 20, 85), ("Sara", 22, 90)]
Output:
[
  {"name": "Ali", "age": 20, "score": 85},
  {"name": "Sara", "age": 22, "score": 90}
]

```python
x = [("Ali", 20, 85), ("Sara", 22, 90)]
l = ["name", "age", "score"]

ans = [dict(zip(l, student)) for student in x]
print(ans)
```
