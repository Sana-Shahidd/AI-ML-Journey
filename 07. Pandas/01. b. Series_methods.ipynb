Drop Duplicates

```python
#this drop the duplicates
temp=pd.Series([1,2,2,1,3,4,3,4,5,6])
temp.drop_duplicates(keep="last") #by default first 
temp.duplicated().sum() #bta deta ha kitni dupliacte values hn
```

isnull

```python
# missing values
# count->non null values
# size-> all values
temp=pd.Series([1,2,np.nan,1,np.nan,4,3,4,np.nan,6])
temp.size #10
temp.count #7
vk.isnull().sum() #data ke andar kiti missing values hn
```

Handling the missing values

```python
# ya to remove->dropna
temp.dropna()
# ya fill->fillna
temp.fillna(0)
```
