```python
# Q-1: Write a program to create an empty series.
import pandas as pd
empty_series=pd.Series([])
print(empty_series)
print(type(empty_series))
```

```python
# Q-2: Write a Pandas program to add, subtract, multiple and divide two Pandas Series.
import numpy as np
first_series=pd.Series([1,2,3])
second_series=pd.Series([1,2,3,5,6])
# np.sum((first_series, second_series)) sum tries to calculate entire sum together
print("sum",first_series + second_series)
print("sub",first_series - second_series)
print("mul",first_series * second_series)
print("div",first_series // second_series)
#using in built
print(first_series.add(second_series, fill_value=0))
print(first_series.sub(second_series, fill_value=0))
print(first_series.mul(second_series, fill_value=0))
print(first_series.div(second_series, fill_value=0))
```

```python
# Q-3 Write a Pandas program to compare the elements of the two Pandas Series.
# Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 10]
first_series=pd.Series([2, 4, 6, 8, 10])
second_series=pd.Series([1, 3, 5, 7, 10])
print("Equal:")
print(first_series == second_series)

print("\nGreater than:")
print(first_series > second_series)

print("\nLess than:")
print(first_series < second_series)
```

```python
# Q-5.Write a function to change the data type of given a column or a Series. Function takes series and data type as input, returns the converted series.
# series = pd.Series([1,2,'Python', 2.0, True, 100])
# change to float type data
series = pd.Series([1,2,'Python', 2.0, True, 100])
def change_dtype(series, dtype):
    return series.astype(dtype)
print(change_dtype(series, float))
```

```python
def change_dtype(series, dtype):
    return pd.to_numeric(series, errors='coerce').astype(dtype) #errors='coerce convert to nan

series = pd.Series([1,2,'Python', 2.0, True, 100])

result = change_dtype(series, float)
print(result)
```
