Aggregation & Statistics

```python
# Find mean age.
print("mean",df["age"].mean())
# Find max and min salary.
print("max_salary: ",df["salary"].max())
print("min_salary: ",df["salary"].min())
# Find total experience.
print("total_experience: ",df["experience"].sum())
# Find median rating.
print("median of rating: ",df["rating"].median())
# Find 25th and 75th percentile of salary.
print("25th percentile of salary: ",np.percentile(df["salary"],25))
print("75th percentile of salary: ",np.percentile(df["salary"],75))
# Find standard deviation of age.
print("Standard deviation of age: ",df["age"].std())
# Count number of employees in each department. .size->for countig the number of rows in each department
#value_Count->count the unique values and works only on series not on dataframe
departments=df.groupby("department").size()
print(departments)
# Count unique names.
count_unique_names=df["name"].drop_duplicates().size
print("Count of unique names: ",count_unique_names)
# Find average salary of IT department employees.
avg_salary=(df[df["department"]=="IT"].salary).mean()
print("average salary of it department: ",avg_salary)
# Find the sum of salaries for employees older than 50.
salaries=df[df["age"]>50]["salary"].sum()
print("sum of salaries for employees older than 50: ",salaries)
```

Sorting & Ranking

```python
# Sort salaries in descending order.
salary=df["salary"].sort_values(ascending=False)
print("salary: \n",salary)
# Get top 10 highest salaries.
print("Top 10 salaries: \n",salary.head(10))
# Sort employees by age.
df_sorted_by_age = df.sort_values("age")
print("Sort by age: \n",df_sorted_by_age)
# Rank employees by rating (highest rating = rank 1).
rate_by_rating=df.sort_values("rating", ascending=False)
print("Sort by rating: \n",rate_by_rating)
# Sort names alphabetically.
sort_by_name=df["name"].sort_values()
print("Sort alphabetically: \n",sort_by_name)
```

Arithmetic & Transformations

```python
# Increase all salaries by 10%.
df["salary"]= df["salary"].apply(lambda x: x*0.10)
#second way
df["salary"] = df["salary"] * 1.10
print(df["salary"])
# Calculate salary per year of experience (avoid division by zero).
df["salary_per_experience"] = df["salary"] / df["experience"].replace(0, np.nan)
print(df[["name", "salary", "experience", "salary_per_experience"]])
# Convert age to months. multiply the number of years by 12 and add the remaining months
df["age"]=df["age"]*12
print("Age to Months",df["age"])
# Round rating to nearest integer.
df["rating_rounded"]=np.round(df["rating"], 0).astype(int)
print(df[["rating", "rating_rounded"]])
# Apply a function to double the experience.
df["experience"]= df["experience"].apply(lambda x: x*2)
#other way
df["experience"]= df["experience"]*2
print(df["experience"])
```

String Operations

```python
# Convert all names to uppercase.
df["upper"]=df["name"].str.upper()
print(df["upper"])
# Find employees whose name contains "n".
df[df["name"].str.contains("n", case=False)]
# Count length of each name.
df["name_length"] = df["name"].str.len()
print(df["name_length"])
# Check if names start with a vowel. #list not applicable here
df[df["name"].str.startswith(("a","e","i","o","u","A","E","I","O","U"))]
# Create a Series of name initials. str.strip() removes spaces
df["initials"]=df["name"].str[0]
df["initials"]
```

Missing Values (simulate a few if needed)

```python
# 40.	Check for missing values in salary.
na_values=df[df["salary"].isna()]
# 41.	Fill missing salaries with the mean salary.
df["salary"] = df["salary"].fillna(df["salary"].mean())
# 42.	Drop rows where rating is missing.
df = df[df["rating"].notna()]
# 43.	Replace all ratings below 2.0 with 2.0.
df["rating"] = df["rating"].apply(lambda x: 2.0 if x < 2.0 else x)
df["rating"] = df["rating"].clip(lower=2.0)
df["rating"] 
```

Cumulative & Window Functions

```python
# Compute cumulative sum of salaries.
np.cumsum(df["salary"])
#pandas way
df["salary_cumsum"] = df["salary"].cumsum()
print(df["salary_cumsum"])
# Compute cumulative maximum of ratings.
df["rating_cumrating"] = df["rating"].cummax()
print(df["rating_cumrating"])
# Find rolling mean of salary with window size 5.
df["salary_roll_mean"] = df["salary"].rolling(window=5).mean()
print(df["salary_roll_mean"])
```

Indexing & Selection

```python
# Set id as the index and select the salary of employee with id=500.
df=pd.read_csv("pandas_series_dataset.csv").set_index("id")
salary_500=df.loc[500, "salary"] #label
salary_500
# Reset the index.
df.reset_index(inplace=True)
# Access salary using .iloc[100]. in iloc must pass index/ uses positions, never index labels.
df.iloc[100, 3] #100 row , 3 column
# Access salary using .loc[100]. loc uses labels 
df.loc[100, "salary"]
```

| Method                       | Indexing Type  | Example                        | Notes                     |
| ---------------------------- | -------------- | ------------------------------ | ------------------------- |
| `.loc[row_label, col_label]` | Label-based    | `df.loc[500, "salary"]`        | Uses index label (id=500) |
| `.iloc[row_pos, col_pos]`    | Position-based | `df.iloc[100, 3]`              | 100th row, 4th column     |
| `reset_index()`              | Resets index   | `df.reset_index(inplace=True)` | Makes old index a column  |
