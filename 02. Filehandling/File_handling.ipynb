```python
import json
import re
from collections import Counter

strings = """Alice was beginning to get very tired of sitting by her sister
            on the bank, and of having nothing to do: once or twice she had
            peeped into the book her sister was reading, but it had no
            pictures or conversations in it, `and what is the use of a book,'
            thought Alice `without pictures or conversation?'

            So she was considering in her own mind (as well as she could,
            for the hot day made her feel very sleepy and stupid), whether
            the pleasure of making a daisy-chain would be worth the trouble
            of getting up and picking the daisies, when suddenly a White
            Rabbit with pink eyes ran close by her.

            There was nothing so VERY remarkable in that; nor did Alice
            think it so VERY much out of the way to hear the Rabbit say to
            itself, `Oh dear! Oh dear! I shall be late!'  (when she thought
            it over afterwards, it occurred to her that she ought to have
            wondered at this, but at the time it all seemed quite natural);
            but when the Rabbit actually TOOK A WATCH OUT OF ITS WAISTCOAT-
            POCKET, and looked at it, and then hurried on, Alice started to
            her feet, for it flashed across her mind that she had never
            before seen a rabbit with either a waistcoat-pocket, or a watch to
            take out of it, and burning with curiosity, she ran across the
            field after it, and fortunately was just in time to see it pop
            down a large rabbit-hole under the hedge."""

word_list = ['alice', 'wonder', 'natural']

# Normalize text (lowercase + remove punctuation)
clean_text = re.findall(r'\b\w+\b', strings.lower())

# Create frequency dictionary
word_freq = Counter(clean_text)
serialized_dict = json.dumps(word_freq)
# Load serialized dict back (optional demonstration)
deserialized_dict = json.loads(serialized_dict)
deserialized_dict
# Get counts for given word_list
result = {word: deserialized_dict.get(word, 0) for word in word_list}
print("Serialized Dictionary:")
print(serialized_dict)
print("\nWord Counts from Given List:")
print(result)
```

```python
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

print(gcd(16, 24)) 
```

```python
def edit_distance(s, t):
    if len(s) == 0:
        return len(t)
    if len(t) == 0:
        return len(s)

    cost = 0
    if s[-1] != t[-1]:
        cost = 1

    d1 = edit_distance(s[:-1], t) + 1
    d2 = edit_distance(s, t[:-1]) + 1
    d3 = edit_distance(s[:-1], t[:-1]) + cost

    return min(d1, d2, d3)

# Example
print(edit_distance("kitten", "sitting"))  # Output: 3
```

```python
def run_length_encode(data):
    if len(data) == 0:
        return []

    def helper(data, index, current, count, result):
        if index == len(data):
            result.extend([current, count])
            return result

        if data[index] == current:
            return helper(data, index + 1, current, count + 1, result)
        else:
            result.extend([current, count])
            return helper(data, index + 1, data[index], 1, result)

    return helper(data, 1, data[0], 1, [])

# Main Program
user_input = input("Enter a string: ")
compressed = run_length_encode(user_input)
print(compressed)
```

```python
def decimal_to_binary(n):
    if n == 0:
        return "0"
    if n == 1:
        return "1"
    return decimal_to_binary(n // 2) + str(n % 2)

# Example
print(decimal_to_binary(10))  # Output: 1010
```

```python
# JSON cannot directly serialize custom class objects.
# You must convert the object into a dictionary first.
import json
class Car:
    def __init__(self, brand, model, year, price):
        self.brand = brand
        self.model = model
        self.year = year
        self.price = price

    def to_dic(self):
            return{
                "brand": self.brand,
                "model": self.model,
                "year": self.year,
                "price": self.price
            }
car1 = Car("Toyota", "Corolla", 2022, 4500000)

with open("car.json", "w") as f:
    json.dump(car1.to_dic(), f, indent=4)
```

```python
import pickle

data = {"name": "Ali", "age": 22}

# Pickle (save)
with open("data.pkl", "wb") as f:
    pickle.dump(data, f)

# Unpickle (load)
with open("data.pkl", "rb") as f:
    new_data = pickle.load(f)

print(new_data)
```
